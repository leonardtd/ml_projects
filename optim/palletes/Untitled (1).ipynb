```
df_query['rounded_heightcm'] = df_query['heightcm'].round(0)  # round to the nearest integer, adjust as needed

# Group by the rounded height
grouped = df_query.groupby('rounded_heightcm')

# Now you can loop through each group (each unique rounded height)
for name, group in grouped:
    print(f"Height: {name}")
    print(f"Case Numbers: {group['Case Number'].tolist()}")
    print("----")
```

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assume df_query is your DataFrame after reading from 'df_pallet_detail.csv'
# df_query = pd.read_csv('df_pallet_detail.csv')

# Plotting the histogram
plt.figure(figsize=(12, 8))  # Set figure size
n, bins, patches = plt.hist(df_query['heightcm'], bins=20, color='blue', edgecolor='black')

# Annotate each bin
for i in range(len(patches)):
    plt.annotate(str(int(n[i])), (patches[i].get_x() + patches[i].get_width() / 2, n[i]),
                 ha='center', va='bottom')

plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Case Numbers by Height')
plt.grid(True)

# Show the plot
plt.show()
```
